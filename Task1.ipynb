```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head(4)
```

```python
data.info()
```

```python
data.columns
```

```python
data['smoker'].value_counts().plot(kind='bar', rot = 0, color = 'blue')
plt.text(x = 0.98, y = data.smoker.value_counts()[1]+1, s = data.smoker.value_counts()[1])
plt.text(x = -0.05, y = data.smoker.value_counts()[0]+2, s = data.smoker.value_counts()[0])

plt.title('Bar Plot')
plt.xlabel('Smoker')
plt.ylabel('Count')

plt.show()
```

```python

plt.hist(data['tip'], bins=5)
plt.title("Histogram Plot")
plt.xlabel("tip")
plt.ylabel("frequency")
plt.show()
```

```python
sns.displot(data['total_bill'])
plt.show()
```

```python
sns.countplot(x='day', data=data, hue='time')
plt.show()
```

```python

numeric_df = data.loc[:, ['total_bill', 'tip', 'size']]
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True )
```
